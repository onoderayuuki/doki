```python
from IPython.display import display,Image, HTML
# html表示
def show_files(folder_path):
    display(Image(filename=folder_path+'.png'))
    for root, dirs, files in sorted(os.walk(folder_path)):
        for file in sorted(files):
            if file.endswith(".html"):
                file_path = os.path.join(root, file)
                print(file_path)
                display(HTML(filename=file_path))
```

```python
def get_directory_structure(root_directory):
    directory_structure = {}

    for dirpath, dirnames, filenames in os.walk(root_directory):
        if dirpath == root_directory:
            continue

        parent_directory = os.path.relpath(dirpath, root_directory)
        files = sorted(filenames)  # ファイル名を名前順にソートする

        directory_structure[parent_directory] = files

    return directory_structure


root_directory = '/content/drive/MyDrive/PointCloudData/土器実測自動化/docs/'
my_dir = get_directory_structure(root_directory)

# print(my_dir)
```

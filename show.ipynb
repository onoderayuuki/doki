```python
from IPython.display import display,Image, HTML
# html表示
def show_files(folder_path):
    display(Image(filename=folder_path+'.png'))
    for root, dirs, files in sorted(os.walk(folder_path)):
        for file in sorted(files):
            if file.endswith(".html"):
                file_path = os.path.join(root, file)
                print(file_path)
                display(HTML(filename=file_path))
```

```python
def get_folder_names(path):
    folder_names = []
    for root, dirs, files in os.walk(path):
        for dir_name in dirs:
            folder_names.append(path+'/'+dir_name)
    folder_names.sort()  # フォルダ名を名前順にソート
    return folder_names
```

```python
# パスを指定してフォルダ名のリストを取得
root_path = '/content/drive/MyDrive/PointCloudData/土器実測自動化'
folders_list = get_folder_names(root_path)
print(folders_list)
```

```python
def read_html_files(folders_list,file_name):
  html_contents = []
  for path in folders_list:
    try:
      with open(path+'/'+file_name, 'r') as file:
          content = file.read()
          html_contents.append(content)
    except:
      pass
  return html_contents


html_0_pcd_file = read_html_files(folders_list,'0_pcd_file.html')
```
